```python
import pandas as pd

# Replace 'your_file.csv' with the path to your CSV file
df = pd.read_csv('C:/Users/isu_v/Desktop/langchain/image_description/sandbox_img_selection/sandbox_imgs_details.csv')

# Initialize a Counter for keys
key_counter = Counter()

# Iterate over the rows and dictionaries in the third column
for index, row in df.iterrows():
    dictionary = eval(row.iloc[2])  # Assuming the dictionaries are in the third column
    for key in dictionary.keys():
        key_counter[key] += 1

# Find the most common keys
most_common_keys = key_counter.most_common()

# Find the least common keys
least_common_keys = key_counter.most_common()[-5:]  # Change 5 to the number of least common keys you want to find

# Print the results
print("Most common keys:")
for key, count in most_common_keys:
    print(f"{key}: {count} times")

print("\nLeast common keys:")
for key, count in least_common_keys:
    print(f"{key}: {count} times")
# In this script, we first read the CSV file using Pandas, assuming the dictionaries are in the third column. We then iterate over the rows, extract and evaluate the dictionaries, and count the keys using a Counter. Finally, we print the most common and least common keys. Make sure the format of the dictionaries in your CSV is valid for this code to work properly.





```
