```python
import numpy as np
import matplotlib.pyplot as plt


with open(r'outputs_perception/liveliness.csv', 'r', encoding='utf-8-sig') as file:
    content = file.read().strip().split(';')
    values = np.array([float(value) for value in content if value])

# Reshape the array to a 11x16 grid 
num_agents = 11
grid_size = 16
values = values.reshape((num_agents, grid_size))

# Create a figure and axis
fig, ax = plt.subplots()

# Create a grid of squares and color them based on the values
cax = ax.matshow(values, cmap='RdYlGn', aspect='auto')

# Set the labels for the axes without ticks
ax.set_xticks(np.arange(grid_size))
ax.set_yticks(np.arange(num_agents))  # Adjusted the number of ticks for rows only

# Update the x-axis tick labels to use letters from 'A' to 'P'
ax.set_xticklabels([chr(65 + i) for i in range(grid_size)])
ax.set_yticklabels([str(i) if i < num_agents else 'Avg' for i in range(1, num_agents + 1)])

# Set labels on the left side and top of the graph
ax.set_ylabel('Agents', fontsize=10)
ax.set_xlabel('Images', fontsize=10)

# Remove minor ticks
ax.tick_params(axis='both', which='major', size=0)
ax.set_xticks([], minor=True)
ax.set_yticks([], minor=True)

# Remove y-axis label
ax.yaxis.set_label_coords(-0.07, 0.5)

# Move x-axis label to the top
ax.xaxis.set_label_coords(0.5, 1.12)

# Set the title at the bottom
ax.set_title('Liveliness', fontsize=14, y=-0.15)

# Add text annotations for each value
for i in range(num_agents):
    for j in range(grid_size):
        if i == num_agents - 1:  # Check if it's the last row
            ax.text(j, i, f'{values[i, j]:.1f}', ha='center', va='center', color='black')
        else:
            ax.text(j, i, f'{int(values[i, j])}', ha='center', va='center', color='black')

# Add colorbar
cbar = fig.colorbar(cax, ax=ax, pad=0.02)
cbar.set_label('Values', rotation=270, labelpad=10)

# Show the plot
plt.show()
```
